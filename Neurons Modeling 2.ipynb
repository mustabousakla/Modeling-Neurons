```python
#CON SCIPY MEJOR



C_m = 9*pi
E_Na =115
E_K =-12
V_rest =10.6
G_Na =1080*pi
G_K =324*pi
G_m =2.7*pi
a_A= 1.1
a_G= 5
b_A=0.19
b_G=0.3
g=10
I=280
E_A=60
E_G=20

def a_m(V):
    return (25 - V)/(10*(exp((25-V)/10)-1))

def b_m(V):
    return 4 * exp(-V/18)

def a_n(V):
    return (10 - V)/(100*(exp((10-V)/10) - 1))
def b_n(V):
    return 0.125 * exp(-V/80)

def a_h(V):
    return 0.07 * exp(-V/20)
def b_h(V):
    return 1/(exp((30 - V)/10) + 1)

def f(V):
    return 1/(1+exp((62-V)/5))

def I_i(g,r_i,E_i,V):
    return g*r_i*(E_i-V)


def ENunidir(y,t):
    V,m,n,h,r_A,V2,m2,n2,h2=y
    dydt=[]
    dydt.append((G_Na *m**3 * h* (E_Na - V) + G_K * n**4 * (E_K - V) + G_m * (V_rest - V) + I)/C_m)
    dydt.append(a_m(V)*(1-m)-b_m(V)*m)
    dydt.append(a_n(V)*(1-n)-b_n(V)*n)
    dydt.append(a_h(V)*(1-h)-b_h(V)*h)
    dydt.append(a_A*f(V)*(1-r_A)-b_A*r_A)
    dydt.append((G_Na *m2**3 * h2* (E_Na - V2) + G_K * n2**4 * (E_K - V2) + G_m * (V_rest - V2) + I+I_i(g,r_A,E_A,V2))/C_m)
    dydt.append(a_m(V2)*(1-m2)-b_m(V2)*m2)
    dydt.append(a_n(V2)*(1-n2)-b_n(V2)*n2)
    dydt.append(a_h(V2)*(1-h2)-b_h(V2)*h2)

    return dydt



y0=[V_rest, np.random.uniform(),np.random.uniform(),np.random.uniform(),np.random.uniform(), V_rest,
   np.random.uniform(),np.random.uniform(),np.random.uniform()]
t=np.linspace(0,200,10000)

sol=odeint(ENunidir,y0,t)

print('feta la integració')

plt.plot(t, sol[:, 0], 'b',label='presynaptic')
plt.xlabel('t (ms)')
plt.ylabel('V (mV)')
plt.ylim(-20,120)
plt.title('E-->N')
plt.plot(t, sol[:, 5],'r', label='postsynaptic')
plt.legend()
```
